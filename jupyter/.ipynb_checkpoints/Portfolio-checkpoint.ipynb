```python
XIU = pd.read_csv('./assets/XIU.csv')
XIU = XIU.iloc[::-1]
XIU['Date']= pd.to_datetime(XIU['Date'])
XIU.set_index('Date', inplace=True)
```

```python
CGL = pd.read_csv('./assets/CGL.csv')
CGL = CGL.iloc[::-1]
CGL['Date']= pd.to_datetime(CGL['Date'])
CGL.set_index('Date', inplace=True)
```

```python
ZDV = pd.read_csv('./assets/ZDV.csv')
ZDV = ZDV.iloc[::-1]
ZDV['Date']= pd.to_datetime(ZDV['Date'])
ZDV.set_index('Date', inplace=True)
```

```python
prices = pd.DataFrame(XIU['Price']).join(GLD['Adj Close'])
prices.columns=['XIU', 'GLD']
prices = prices.join(QQQ['Adj Close'])
prices.columns=['XIU', 'GLD', 'QQQ']
prices = prices.join(SPY['Adj Close'])
prices.columns=['XIU', 'GLD', 'QQQ', 'SPY']
prices = prices.join(GM['Adj Close'])
prices.columns=['XIU', 'GLD', 'QQQ', 'SPY', 'GM']
prices = prices.join(USO['Adj Close'])
prices.columns=['XIU', 'GLD', 'QQQ', 'SPY', 'GM', 'USO']
prices = prices.join(ZDV['Price'])
prices.columns=['XIU', 'GLD', 'QQQ', 'SPY', 'GM', 'USO', 'ZDV']
prices = prices.join(CGL['Price'])
prices.columns=['XIU', 'GLD', 'QQQ', 'SPY', 'GM', 'USO', 'ZDV', 'CGL']
```

```python
prices = prices[['XIU', 'GLD', 'QQQ', 'SPY', 'USO', 'ZDV']]
# prices = prices[prices.index.year > 2016]
```

```python
## ANNUAL RETURNS
prices.fillna(method='bfill', inplace=True)
YTDreturn = prices.groupby(pd.TimeGrouper('A')).transform(lambda x: x/x.iloc[0]-1.0)
YTDReturns = []
for year in YTDreturn.index.year.unique():
       YTDReturns.append( (YTDreturn[YTDreturn.index.year ==year].iloc[-1]))
yearly_returns = pd.DataFrame(YTDReturns)

    
## MONTHLY RETURNS
MNreturn = prices.groupby(pd.TimeGrouper('M')).transform(lambda x: x/x.iloc[0]-1.0)
MNTReturns = []
for year in MNreturn.index.year.unique():
        for month in MNreturn[MNreturn.index.year==year].index.month.unique():
            MNTReturns.append( (MNreturn[(MNreturn.index.year ==year) & (MNreturn.index.month==month)].iloc[-1])) 
monthly_returns = pd.DataFrame(MNTReturns)


## DAILY RETURNS
daily_returns=pd.DataFrame(prices[stocks[0]].pct_change())
for stock in stocks[1:]:
    daily_returns = daily_returns.join(prices[stock].pct_change())
 
## STDS
monthly_stds = prices.groupby([prices.index.year, prices.index.month]).std()
yearly_stds = prices.groupby(prices.index.year).std()
```

```python
## 'Kash Ratio'
# sns.heatmap(pd.DataFrame(yearly_returns.values/ yearly_stds.values, columns=stocks))
pd.DataFrame(yearly_returns.values/ yearly_stds.values, columns=stocks).mean().sort_values(ascending=False).plot(kind='bar')
```

```python
mean_daily_returns = daily_returns.mean()
cov_matrix = daily_returns.cov()
num_portfolios = 100000


results = np.zeros((3,num_portfolios))
bestSharpe = 0
bestAlloc = []

for i in range(1, num_portfolios):
    weights = np.random.random(len(stocks))
    weights /= np.sum(weights)
    portfolio_return = np.sum(mean_daily_returns * weights) * 252
    portfolio_std_dev = np.sqrt(np.dot(weights.T,np.dot(cov_matrix, weights))) * np.sqrt(252)
    sharpe = portfolio_return / portfolio_std_dev
    results[0,i] = portfolio_return
    results[1,i] = portfolio_std_dev
    results[2,i] = sharpe
    if sharpe > bestSharpe:
            bestSharpe = sharpe;
            bestAlloc=weights;
    
#convert results array to Pandas DataFrame
results_frame = pd.DataFrame(results.T,columns=['return','stdev','sharpe'])
 
#locate position of portfolio with highest Sharpe Ratio
max_sharpe_port = results_frame.iloc[results_frame['sharpe'].idxmax()]
#locate positon of portfolio with minimum standard deviation
min_vol_port = results_frame.iloc[results_frame['stdev'].idxmin()]
plt.figure( figsize=(20,8))
                             
plt.scatter(results_frame.stdev,results_frame['return'],c=results_frame.sharpe,cmap='RdYlBu', s=10)
plt.xlabel('Volatility')
plt.ylabel('Returns')
plt.colorbar()
plt.scatter(max_sharpe_port[1],max_sharpe_port[0],marker=(5,1,0),color='r',s=100)

td = pd.DataFrame(bestAlloc*100, index=stocks, columns=['Sharpe'])
print(bestSharpe)
td.sort_values(by="Sharpe", ascending=False)
```
